```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
df_data=pd.read_csv("orange_small_churn_data.csv")
name=['target']
target=pd.read_csv("orange_small_churn_labels.csv",names=name)
df_data.head()
```

```python
%pylab inline
```

```python
to_drop=[]
for i in df_data.columns:
    if df_data.isna().sum()[i]/df_data.shape[0]==1:
#         print(i,df_data.isna().sum()[i]/df_data.shape[0])
        to_drop.append(i)
df_data.drop(columns=to_drop,inplace=True)
df_data.shape
```

__First approach__

```python
n=0
to_drop_90=[]
for i in df_data.columns:
    if df_data.isna().sum()[i]/df_data.shape[0]>0.9:
        n+=1
        to_drop_90.append(i)
#         print(i,df_data.isna().sum()[i]/df_data.shape[0])
print(f'absolute :{n}, persent: {n/(df_data.shape[1])*100} %')
```

```python
df_data.drop(columns=to_drop_90,inplace=True)
df_data.shape
```

```python
df_data.info()
```

```python
from sklearn.preprocessing import StandardScaler
df_data.loc[:,'Var6':'Var189'] = df_data.loc[:,'Var6':'Var189'].fillna(0)
scaler = StandardScaler()
# transform data
df_data.loc[:,'Var6':'Var189'] = scaler.fit_transform(df_data.loc[:,'Var6':'Var189'])
df_data.head()
```

```python
df_data.loc[:,'Var192':'Var229']=df_data.loc[:,'Var192':'Var229'].fillna(-100)
```

```python
categ=[]
for i in df_data.loc[:,'Var192':'Var229'].columns:
    categ.append(i)
```

```python
dummy=pd.get_dummies(df_data.loc[:,'Var192':'Var229'])
df_data_no_categ=df_data.drop(columns=categ)
```

```python
df_new=pd.concat([df_data_no_categ,dummy],axis=1)
df_new.head()
```

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.model_selection import cross_val_score
clf = RidgeClassifier()
print(cross_val_score(clf, df_data_no_categ, target, cv=3,scoring='roc_auc').mean())
# clf = RidgeClassifier().fit(X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
print(cross_val_score(clf, df_data_no_categ, target, cv=3,scoring='roc_auc').mean())
# clf = RidgeClassifier().fit(X, y)
```

```python
import xgboost as xgb
xg_reg = xgb.XGBClassifier()
print(cross_val_score(xg_reg, df_data_no_categ, target, cv=5,scoring='roc_auc').mean())
# cv=3 ->0.7195217492152777
```
